```python
from sklearn.manifold import LocallyLinearEmbedding
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
import numpy as np
%matplotlib widget
```

```python
filepath_HCC_Fil = "./raw_data/HCC1806_SmartS_Filtered_Data.txt"
df_HCC_Fil = pd.read_csv(filepath_HCC_Fil,delimiter="\ ",engine='python',index_col=0)
print("Dataframe dimensions:", np.shape(df_HCC_Fil))

filepath_HCC_Train = "./raw_data/HCC1806_SmartS_Filtered_Normalised_3000_Data_train.txt"
df_HCC_Train = pd.read_csv(filepath_HCC_Train,delimiter="\ ",engine='python',index_col=0)
print("Dataframe dimensions:", np.shape(df_HCC_Train))

filepath_HCC_Test = "./raw_data/HCC1806_SmartS_Filtered_Normalised_3000_Data_test_anonim.txt"
df_HCC_Test = pd.read_csv(filepath_HCC_Test,delimiter="\ ",engine='python',index_col=0)
print("Dataframe dimensions:", np.shape(df_HCC_Test))
```

```python
lle1 = LocallyLinearEmbedding(n_components=2, n_neighbors=10)
X_reduced1 = lle1.fit_transform(df_HCC_Train)

X_df1 = pd.DataFrame(X_reduced1, columns=["PC1","PC2"])
print(X_df1.head())
sns.scatterplot(data = X_df1, x="PC1", y="PC2")
```

```python
lle2 = LocallyLinearEmbedding(n_components=3, n_neighbors=10)
X_reduced2 = lle2.fit_transform(df_HCC_Train)

X_df2 = pd.DataFrame(X_reduced2, columns=["LLE1","LLE2","LLE3"])
```

```python
sns.set(style = "darkgrid")

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = X_df2["LLE1"]
y = X_df2["LLE2"]
z = X_df2["LLE3"]

ax.set_xlabel("LLE1")
ax.set_ylabel("LLE2")
ax.set_zlabel("LLE3")

ax.scatter(x, y, z)

plt.show()
```
